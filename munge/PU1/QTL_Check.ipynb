```r
# Correct for the multiple phenotypes being tested
library(qvalue) 
d = read.table("processed/PU1/qtltools/output/cqn/PU1.permuted.txt.gz", head=FALSE, stringsAsFactors=FALSE)
d=d[!is.na(d$V20),]
d$qval=qvalue(d$V20)$qvalue
write.table(d[which(d$qval <= 0.05), ], "processed/PU1/qtltools/output/cqn/results.genes.significant.txt", quote=FALSE, row.names=FALSE, col.names=FALSE)
```
